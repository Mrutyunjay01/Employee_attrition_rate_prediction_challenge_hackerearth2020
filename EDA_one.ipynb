```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
# construct the Data Matrix for XgBdoost
dm = xgb.DMatrix(data=train_df.drop(['Attrition_rate'], axis=1), label=train_df['Attrition_rate'])
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
xgbreg = xgb.sklearn.XGBRegressor()
parameters = {'nthread':[4], #when use hyperthread, xgboost may become slower
              'objective':['reg:linear'],
              'learning_rate': [.03, 0.05, .07], #so called `eta` value
              'max_depth': [5, 6, 7],
              'min_child_weight': [4],
              'silent': [1],
              'subsample': [0.7],
              'colsample_bytree': [0.7],
              'n_estimators': [500]}

xgb_grid = GridSearchCV(xgbreg,
                        parameters,
                        cv = 2,
                        n_jobs = 5,
                        verbose=True)

xgb_grid.fit(train_df.drop(['Attrition_rate'], axis=1), train_df['Attrition_rate'])
```

```python
xgb_grid.best_score_
```

```python
xgb_grid.best_estimator_
```

```python
xgb_grid.best_params_
```

```python
params_updated = xgb_grid.best_params_
cv_results = xgb.cv(dtrain=dm, params=params_updated, nfold=3,
                    num_boost_round=50,early_stopping_rounds=10,metrics="rmse", as_pandas=True, seed=123)
```

```python
cv_results.head()
```

```python
pred = xgb_grid.predict(X=train_df.drop(['Attrition_rate'], axis=1))
```

```python
pred
```

```python
train_df['Attrition_rate'].values
```

```python
np.sqrt(mean_squared_error(y_pred=pred, y_true=train_df['Attrition_rate'].values))
```

```python
test_df[['Age', 'Time_of_service', 'Pay_Scale', 'Work_Life_balance', 'VAR2', 'VAR4']] = test_df[['Age', 'Time_of_service', 'Pay_Scale', 'Work_Life_balance', 'VAR2', 'VAR4']].fillna(0)
```

```python
test_df[test_df.columns].isna().sum()
```

```python
(test_df[test_df.columns]==0).sum()
```

```python
test_df_zeros = test_df.iloc[:]
(test_df_zeros[:]==0).sum()
```

```python
test_df['Age'] = test_df['Age'].replace(0, test_df.Age.median())
test_df['Time_of_service'] = test_df['Time_of_service'].replace(0, test_df.Time_of_service.median())
test_df['Time_since_promotion'] = test_df['Time_since_promotion'].replace(0, test_df.Time_since_promotion.median())
test_df['Travel_Rate'] = test_df['Travel_Rate'].replace(0, test_df.Travel_Rate.median())
test_df['VAR2'] = test_df['VAR2'].replace(0, test_df.VAR2.median())
test_df['VAR4'] = test_df['VAR4'].replace(0, test_df.VAR4.median())
test_df['Pay_Scale'] = test_df['Pay_Scale'].replace(0, test_df.Pay_Scale.median())
test_df['Work_Life_balance'] = test_df['Work_Life_balance'].replace(0, test_df.Work_Life_balance.median())
(test_df[test_df.columns]==0).sum()
```

```python
(test_df_zeros[:]==0).sum()
```
